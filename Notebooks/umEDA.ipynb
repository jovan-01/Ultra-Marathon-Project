# Exploratory Data Analysis

### Import Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('ultra_northamerica_classic.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.countplot(data=df, x='Year of event', palette='viridis')
plt.title('Number of Ultra-Marathon Race Records per Year (1923–2022)', fontsize=14)
plt.xlabel('Year of Event')
plt.ylabel('Number of Records')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.countplot(data=df, x='Year of event', hue='Event distance/length', palette='Set2')
plt.title('Number of Races per Year by Distance')
plt.xlabel('Year')
plt.ylabel('Number of Records')
plt.legend(title='Race Distance')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df.drop(703266, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df['Athlete average speed'].describe()
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data=df, x='Year of event', y='Athlete average speed', hue='Event distance/length')
plt.title('Average Speed by Year and Race Distance')
plt.xlabel('Year')
plt.ylabel('Average Speed (km/h)')
plt.legend(title='Distance')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Athlete gender', y='Athlete Age', hue='Event distance/length', split=True, inner = None)
plt.title('Age Distribution by Gender and Distance')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Athlete gender', y='Athlete average speed', hue='Event distance/length', split=True, inner = None)
plt.title('Speed Distribution by Gender and Distance')
plt.xlabel('Gender')
plt.ylabel('Average Speed (km/h)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Athlete country', hue='Event distance/length')
plt.title('Participation by Country and Race Distance')
plt.xlabel('Country')
plt.ylabel('Number of Records')
plt.legend(title='Distance')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Event distance/length', y='Athlete average speed', hue='Athlete country', ci = None)
plt.title('Average Speed by Country and Race Distance')
plt.ylabel('Average Speed (km/h)')
plt.xlabel('Distance')
plt.tight_layout()
plt.show()
```

```python
df['Decade'] = (df['Year of event'] // 10) * 10

plt.figure(figsize=(14, 6))
sns.countplot(data=df, x='Decade', hue='Athlete gender', palette='muted')
plt.title('Gender Distribution of Ultra Runners per Decade (2000s–2020s)')
plt.xlabel('Decade')
plt.ylabel('Number of Records')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
sns.lmplot(data=df, x='Athlete Age', y='Athlete average speed', hue='Event distance/length', lowess=True, height=6, aspect=1.5)
plt.title('Avg Speed vs Age by Distance')
```
